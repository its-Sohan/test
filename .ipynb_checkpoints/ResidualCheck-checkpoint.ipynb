```python
# Importing libs in a separate Cell
import cv2
import numpy as np
from skimage.restoration import denoise_bilateral
import matplotlib.pyplot as plt

#Ends here
```

```python
# Image loading and error check
img = cv2.imread("79ve54pbhadf1.jpeg", cv2.IMREAD_GRAYSCALE)
if img is None:
    raise FileNotFoundError("Image not found or failed to load.")
```

```python
# Normalize to [0, 1] (Floatng point value)32
img_norm = img / 255.0
# Denoise Stick to float until here 
denoised = denoise_bilateral(img_norm, sigma_color=0.3, sigma_spatial=18, channel_axis=None)
# converts to int array          i know its AI and already smootheed so no further neeed
# i  nt required here Memory Efficiency
#    float32 takes 4 bytes per pixel.
#    uint8 takes just 1 byte.
denoised_8bit = (denoised * 255).astype(np.uint8) #denoisedd 8bit int
```

```python
# residual
residual = img - denoised_8bit
# Convert to float32 for FFT
residual = residual.astype(np.float32)
```

```python
# A
f = np.fft.fft2(residual)
fshift = np.fft.fftshift(f).real #Inverse fast fourier trasform
# Compute magnitude
magnitude = np.abs(fshift)

#  Use logarithmic scale for visibility
magnitude_log = 20 * np.log(magnitude + 1)  # log(1 + x) for stability
```

```python
# Show all
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1, 3, 2)
plt.imshow(denoised_8bit, cmap='gray')
plt.title("Denoised")

plt.subplot(1, 3, 3)
plt.imshow(residual, cmap='gray')
plt.title("Noise Residual")
plt.tight_layout()


plt.figure(figsize=(6, 6))
plt.imshow(magnitude_log, cmap='gray')
plt.title("FFT Magnitude Spectrum (Residual)")
plt.show()
```
